```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from scipy.stats import pearsonr
from statsmodels.sandbox.stats.multicomp import multipletests 
import math
```

```python
df = pd.read_csv("gene_high_throughput_sequencing.csv", sep=',')
```

```python
ttest_data=[]
normal = df[df.Diagnosis=="normal"]
early_neoplasia=df[df.Diagnosis=="early neoplasia"]
cancer = df[df.Diagnosis=="cancer"]
k1=0
corr_data1=[]
for i in range(len(df.columns)-2):
    pvalue = scipy.stats.ttest_ind(normal[df.columns[i+2]], early_neoplasia[df.columns[i+2]], equal_var = False)
    mean_c=np.mean(normal[df.columns[i+2]])
    mean_t=np.mean(early_neoplasia[df.columns[i+2]])
    corr_data1.append([df.columns[i+2],mean_c,mean_t, pvalue[1]])
    if (pvalue[1]<0.05):
        k1+=1
print k1
k2=0
corr_data2=[]
for i in range(len(df.columns)-2):
    pvalue = scipy.stats.ttest_ind(early_neoplasia[df.columns[i+2]], cancer[df.columns[i+2]], equal_var = False)
    mean_c=np.mean(early_neoplasia[df.columns[i+2]])
    mean_t=np.mean(cancer[df.columns[i+2]])
    corr_data2.append([df.columns[i+2],mean_c,mean_t, pvalue[1]])
    if (pvalue[1]<0.05):
        k2+=1
print k2
print k1+k2
```

```python
sales_correlation1 = pd.DataFrame.from_records(corr_data1)
sales_correlation1.columns = ['gene', 'mean_c', 'mean_t', 'p']
```

```python
sales_correlation2 = pd.DataFrame.from_records(corr_data2)
sales_correlation2.columns = ['gene', 'mean_c', 'mean_t', 'p']
```

```python
def write_answer_1(k1):
    with open("ans1.txt", "w") as fout:
        fout.write(str(k1))
def write_answer_2(k2):
    with open("ans2.txt", "w") as fout:
        fout.write(str(k2))
write_answer_1(k1)
write_answer_2(k2)
```

```python
import statsmodels.stats.multitest as smm
from statsmodels.sandbox.stats.multicomp import multipletests 
```

```python
reject1, p_corrected1, a1, a2 = multipletests(sales_correlation1.p, 
                                            alpha = 0.025, 
                                            method = 'holm') 
reject2, p_corrected2, b1, b2 = multipletests(sales_correlation2.p, 
                                            alpha = 0.025, 
                                            method = 'holm') 
```

```python
sales_correlation1['p_corrected'] = p_corrected1
sales_correlation1['reject'] = reject1
sales_correlation2['p_corrected'] = p_corrected2
sales_correlation2['reject'] = reject2
print sales_correlation1
print sales_correlation1.mean_t[0]
```

```python
k3=0
fold_change=0
sales_correlation1.reject.value_counts()
for i in range(len(sales_correlation1)):
    if (sales_correlation1.p_corrected[i]<0.025):
        t = sales_correlation1.mean_t[i]
        c = sales_correlation1.mean_c[i]
        if (t > c):
            fold_change=t/c
        else:
            fold_change=-c/t
        if (math.fabs(fold_change)>1.5):
            k3+=1

k4=0
fold_change=0
sales_correlation2.reject.value_counts()
for i in range(len(sales_correlation2)):
    if (sales_correlation2.p_corrected[i]<0.025):
        t = sales_correlation2.mean_t[i]
        c = sales_correlation2.mean_c[i]
        if (t > c):
            fold_change=t/c
        else:
            fold_change=-c/t
        if (math.fabs(fold_change)>1.5):
            k4+=1
print k3,k4
```
